# Построение графика

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.figure(figsize=(7, 3)) # подготовка зоны для рисования
plt.plot(x, y, c='green', lw=2) # отрисовка
plt.text(3.3, 0, r'$\sin(x)$', fontsize=15, rotation=-60) # изображение текста
plt.xlabel('ось $x$') # подпись оси x
plt.ylabel('ось $y$') # подпись оси y
plt.title('график', family="verdana") # заглавие
plt.xticks(np.linspace(1, 10, 11)) # сетка по x
plt.yticks(np.linspace(-1, 1, 5)) # сетка по y
plt.xlim([0, 10]) # лимиты по оси x
plt.ylim([-1, 1]) # лимиты по оси y
plt.grid(1)
```

# Сохранение

```python
fig = plt.figure(figsize=(7, 3)) # задает размер картинки
for t, i in enumerate(np.linspace(0, 1, 10)): 
    plt.plot(x+3*i, y+i, c=(i, i, 1-i), lw=2, label='график '+str(t))
    
plt.legend(loc=(1.01, 0), frameon=True, ncol=2,
           fancybox=True, framealpha=1, shadow=True, borderpad=1)    

# plt.savefig('pic_subs.png', format='png', bbox_inches='tight') выгружает файл в формат png
# plt.savefig('pic_subs.pdf', format='pdf', bbox_inches='tight') выгружает файл в формат pdf
# plt.savefig('pic_subs.ps', format='ps', bbox_inches='tight') выгружает файл в формат ps
# plt.savefig('pic_subs.eps', format='eps', bbox_inches='tight') выгружает файл в формат eps
# plt.savefig('pic_subs.svg', format='svg', bbox_inches='tight') выгружает файл в формат svg
```

# Несколько графиков на одном рисунке

# Диаграммы рассеивания

```python
rng = np.random.RandomState(0) # формирует случайные значения
n = 15
x = rng.randn(n) # создает размер случайным образом 
y = rng.randn(n) # создает размер случайным образом
colors = np.random.rand(n)
sizes = 1000 * np.random.rand(n)

plt.figure(figsize=(7, 4)) # размер графика
plt.scatter(x, y, c=colors, s=sizes, # создает диаграмму рассеивания
            alpha=0.3,
            cmap='viridis')
plt.colorbar(); # создает шкалу справа от графика
```

# Некоторые другие графики

```python
x = np.array([1, 2, 3, 4, 5, 6, 7])
y = np.array([3, 1, 3, 2, 2, 4, 0])
z = np.array([1, 2, 1, 1, 2, 2, 3])
```

```python
plt.figure(figsize=(7, 3))
plt.errorbar(x, y, # создает график ящик с усами
             yerr=z, # создает усы
             fmt='o', # задает формат графика
             color='black', # цвет кружочка
             ecolor='lightgray', # цвет линий
             elinewidth=5, # ширина линии
             capsize=10); # ширина усов
plt.xlim([0, 8])
plt.ylim([-4, 7])
plt.grid()
```

```python
plt.figure(figsize=(7, 3))
plt.plot(x, y, '-', lw=2, color='blue')
plt.fill_between(x, y-z, y+z, # создает границу выше и ниже основной линии
                 color='blue', # цвет границы
                 alpha=0.1) # прозрачность
plt.xlim([0, 8])
plt.ylim([-4, 7])
```

# Плотность

```python
x1 = np.random.normal(0, 0.8, 100)
x2 = np.random.normal(-2, 1, 100)

#kwargs = dict(bins=20, histtype='stepfilled', alpha=0.5, normed=True)
plt.figure(figsize=(7, 3))
plt.hist(x1, bins=20, alpha=0.5, color='blue');
plt.hist(x2, bins=20, alpha=0.5, color='green');
plt.scatter(x1, 12*np.random.rand(100) - 13, c='blue', alpha=0.5)
plt.scatter(x2, 12*np.random.rand(100) - 13, c='green', alpha=0.5)
```
